#### created a test dataframe 

```python
data = [
    (101, "Mumbai", "Goa", 587),
    (102, "Mumbai", "Bangalore", 985),
    (102, "Mumbai", "Bangalore", 985),
    (103, "Delhi", "Chennai", 2208),
    (104, "Delhi", "Chennai", 2208),
    (105, "Bangalore", "Kolkata", 1868),
    (105, "Bangalore", "Kolkata", 1865)
]
```

#### create a schema and column name

```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType, DoubleType

schema = StructType([
    StructField("id", IntegerType()),
    StructField("source", StringType()),
    StructField("destination", StringType()),
    StructField("distance", DoubleType())])
```

#### create a dataframe

```python
travel_df=spark.createDataFrame(data, schema)
```

####2. Select records that satisfy the follwing criteria
1. Source is Mumbai and destination is Bangalore

Show the following approaches
1. Using a single filter
2. Using two filters
3. Using a col function expression
4. Using a dataframe qualifier expression

```python
travel_df.count()
```

#### using single filter

```python
from pyspark.sql.functions import col

df = travel_df.filter(col("source")=="Mumbai")
df.show()
```

#### usinf two filters

##### 👉 Get all records where:

source is Mumbai
AND distance is greater than 900

```python
from pyspark.sql.functions import col

df = travel_df.filter((col('source')=="Mumbai") & (col('distance')>900))
```

```python
df_1 = travel_df.filter("source='Mumbai' and distance>900")
```

```python
df_1.show()
```

```python
df.show()
```

##### 🔹 Question 2 (AND + OR mix)

👉 Get all records where:

(source is Delhi AND distance is greater than 2000)
OR destination is Kolkata

```python
from pyspark.sql.functions import col
df=travel_df.filter(
    (col('source')=="Delhi") 
    &
    (col('distance')>2000)
    |
    (col("destination")=='Kolkata')
    )
df.show()
```
