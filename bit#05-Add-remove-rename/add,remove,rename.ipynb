```python
from pyspark.sql import SparkSession
# Data (manually defined)
data_list = [
    (100, "Prashant", 45, 45000),
    (101, "Tarun", 36, 33000),
    (102, "David", 48, 28000)
]

# Create Schema
schema = "id int, name string, age short, salary double"


# Create DataFrame

#md This will create a DataFrame that spark automatically infers the schema
#df = spark.createDataFrame(data=data_list)


# This will create a DataFrame the .DF transformation can apply to any DF, it allows you to rename all column at once( but not the recommended way)

#df = spark.createDataFrame(data=data_list.toDF("id","name","age","salary")

sample_df = spark.createDataFrame(data=data_list, schema=schema) #bestway

# Show DataFrame
sample_df.show()
```

##### Add following column to your DataFrame
1. increment 10% of the salary up to 3000 maximum increment 
2. revised_salary:salary + increment

```python
from pyspark.sql.functions import expr
salary_df = (
  sample_df
  .withColumns({
    "increment": expr(" case when salary > 30000 then 3000 else salary * 0.01 end"),
    "revised_salary": expr("salary + increment")
  })
)
```

```python
salary_df.display()
```

####3. Add following columns to your Dataframe
* increment: 10% of the salary up to 3000 maximum increment

Replace the following column in your dataframe
* salary: current salary + increment

```python
from pyspark.sql.functions import expr
salary_df = (
  sample_df
  .withColumn(
    "increment", expr(" case when salary > 30000 then 3000 else salary * 0.01 end"))
    .withColumn("salary",expr("salary + increment")
  )
)
```

```python
salary_df.display()
```

#### 4 Add a bath number(uuid) column to your DF

```python
from pyspark.sql.functions import lit
import uuid
batch_id = str(uuid.uuid4())

salary_batch_df = sample_df.withColumn("batch_id", lit(batch_id))
```

```python
salary_batch_df.display()
```

####4. Rename the dataframe colums as listed below
1. increment - annual_increment
2. salary - incremented_salary

```python
new_salary_df = (
    salary_df.withColumnsRenamed({
        "increment": "annual_increment",
        "salary": "incremented_salary"
    })
)

new_salary_df.display()
```

```python
rename_dict = {
    "increment": "annual_increment",
    "salary": "incremented_salary"
}

new_salary_df = salary_df
for old, new in rename_dict.items():
    new_salary_df = new_salary_df.withColumnRenamed(old, new)
```
