####2. Find the top 3 most delayed flights on 2000-01-16 from AUS to ORD

Collect the results into a list and display the output as the following.
```
    AA flight delayed by 5.0 minutes
    AA flight delayed by 2.0 minutes
    UA flight delayed by 2.0 minutes

````python
from pyspark.sql.functions import col, lit
top_3_df=(flight_time_df.filter(
    (col('FL_DATE') == '2000-01-16')
    &
    (col('origin') == "AUS")
    &
    (col('dest') == "ORD")
)
          .withColumn('Delayed_time',col('arr_time')-(col('CRS_ARR_TIME')))
          .orderBy('Delayed_time', ascending=False)
          .limit(3)

)
````

````python
top_3_df.select("FL_DATE","ORIGIN","DEST","Delayed_time").display()
````

##### 👉 .collect() is a PySpark action that retrieves all rows from a DataFrame and returns them as a list of Row objects in Python (driver memory)

````python
top_3_list_of_rows = top_3_df.collect()
````

````python
top_3_list_of_rows
````

#### .asDict() is a method of a PySpark Row object that converts the Row into a Python dictionary (key-value pairs).

2.2 Make a list of dictionary from a list of rows

````python
top_3_list = [row.asDict() for row in top_3_list_of_rows]
````

print the desire value

````python
for i in top_3_list:
    print(i["OP_CARRIER"] 
          + " flight delayed by "
          + str(i["delayed_arrival"].total_seconds()/60)
          + " minutes")
````
