#### create a Dataframe from a Python list

```python
from datetime import datetime, date

# schema
data_list_schema = "id long, name string, joining_date date, salary double, created_at timestamp"

# data
data_list = [
    (1, "Prashant", date(2018, 1, 1), 924.0, datetime(2022, 1, 1, 9, 0)),
    (2, "Sushant", date(2018, 2, 1), 1260.50, datetime(2022, 1, 2, 11, 0)),
    (3, "David", date(2023, 3, 1), 765.0, datetime(2022, 1, 3, 10, 7))
]

# create dataframe
df = spark.createDataFrame(data_list, data_list_schema)

# df.show()
# df.printSchema()
```
