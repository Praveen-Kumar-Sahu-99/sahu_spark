```python
raw_fire_df = (
    spark.read.format("csv")
    .option('header', 'true')
    .option("inferSchema", "true")
    .load('/Volumes/dev/spark_db/datasets/spark_programming/data/sf-fire-calls.csv')
)
```

```python
display(raw_fire_df)
```

```python
raw_fire_df.count()
```

```python
raw_fire_df.printSchema()
```

```python
from pyspark.sql.functions import try_to_timestamp, col, lit

fire_df = raw_fire_df \
    .withColumn("AvailableDtTm", try_to_timestamp(col("AvailableDtTm"), lit("MM/dd/yyyy hh:mm:ss a"))) \
    .withColumn("Zipcode", col("Zipcode").cast("string")) \
    .withColumn("FinalPriority", col("FinalPriority").cast("string"))

                    
```

```python
fire_df.printSchema()
```

```python
%sql

desc dev.spark_db.sf_fire_calls
```

Save the final df inot the target table

```python
fire_df.write.mode('overwrite').saveAsTable('dev.spark_db.sf_fire_calls')
```

```python
fire_df.filter(fire_df['AvailableDtTm'].isNull() & (raw_fire_df['AvailableDtTm'].isNotNull())).select('AvailableDtTm').show(5, False)
```
