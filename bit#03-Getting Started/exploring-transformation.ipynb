#### what were the most common call type

```python
q4_df = (
    fire_df.select(fire_df.CallType)
                   .groupBy(fire_df.CallType)
                   .count()
                   .orderBy("count", ascending=False)
                   )
```

```python
display(q4_df)
```

```python
q5_df= (
    fire_df.select("CallType","Zipcode")
    .where("CallType is not null")
    .groupBy("CallType","Zipcode")
    .count()
    .orderBy("count", ascending=False)
    .limit(3)

)
```

#### what San Fransiso neighborhood are in the zip code 94102 and 94103

```python
q6_df = (
    fire_df.select("Neighborhood","City")
    .where("ZipCode in (94102,94103)")
    .limit(10)
)
```

```python
display(q6_df)
```

What was the sum of all calls, average, min and max of the response times for calls

```python
from pyspark.sql.functions import min, max, avg,sum
q7_df = (
    fire_df.select("CallType","Delay","NumAlarms")
    .where('CallType is not null')
    .groupBy("CallType")
    .agg(
        sum("NumAlarms").alias("NumAlarms_sum"),
        avg("Delay").alias("Delay_avg"),
        min("Delay").alias("Delay_min"),
        max("Delay").alias("Delay_max")
        )
)
```

```python
display(q7_df)
```

#### How many distint years of data is in the CSV file

```python
from pyspark.sql.functions import to_timestamp,year,col
q8_df = (
    fire_df
    .select(year(to_timestamp("CallDate","MM/dd/yyyy")).alias("year"))
    .distinct()
    .orderBy("year", ascending=False)
)
```

```python
display(q8_df)
```

#### what neighborhoods in san franiso had the worst response time in 2018?

```python
from pyspark.sql.functions import to_timestamp, year

q9_df = (
    fire_df
    .select('Neighborhood',year(to_timestamp("CallDate","MM/dd/yyyy")).alias("year"), "City")
    .where("year = 2018 AND City = 'San Francisco'")
)

q9_df.show()
```

```python
from pyspark.sql.functions import to_timestamp, year, col

q9_df = (
    fire_df
    .select(year(to_timestamp("CallDate","MM/dd/yyyy")).alias("year"), "City")
    .where((col("year") == 2018) & (col("City") == "San Francisco"))
)

q9_df.show()
```
